## Imports

```python
import numpy as np
import pandas as pd
import os
import sys
import shutil
import matplotlib.pyplot as plt
import seaborn as sns
```

## Pathways and Image Directories

```python
data_path = os.path.join('..', 'Data')

images_path = os.path.join(data_path, 'Images')

test_path = os.path.join(data_path, 'test_images')
train_path = os.path.join(data_path, 'train_images')
val_path = os.path.join(data_path, 'val_images')

image_folders = os.listdir(images_path)

test_folders = os.listdir(test_path)
train_folders = os.listdir(train_path)
val_folders = os.listdir(val_path)
```

## Visualizations

```python
image_counts = []
for folder in image_folders:
    path = os.path.join(images_path, folder)
    image_counts.append(len(os.listdir(path)))
```

```python
fig, ax = plt.subplots(figsize=(13,8))
plt.title('Distribution of Dog Breed Image Counts ', fontsize = 25)
ax.tick_params(axis = 'both', labelsize = 15)
plt.xlabel('Image Counts', fontsize = 20)
plt.ylabel('# of Breeds', fontsize = 20)
plt.hist(image_counts);
```

```python
def distribution_maker(folders, path, name):
    dis_image_counts = []
    for folder in folders:
        dis_path = os.path.join(path, folder)
        dis_image_counts.append(len(os.listdir(dis_path)))
    fig, ax = plt.subplots(figsize=(16,8))
    plt.title('Distribution of Dog Breed ' + name + ' Image Counts ', fontsize = 25)
    ax.tick_params(axis = 'both', labelsize = 15)
    plt.xlabel('Image Counts', fontsize = 20)
    plt.ylabel('# of Breeds', fontsize = 20)
    plt.hist(dis_image_counts);
```

```python
distribution_maker(train_folders, train_path, 'Train')
```

```python
distribution_maker(test_folders, test_path, 'Test')
```

```python
distribution_maker(val_folders, val_path, 'Validation')
```
